```python
def pad_img(img):
    a = np.array(img.resize((img.size[0], 28)))
    total_pad = 28 - a.shape[1]
    l_pad = ceil(total_pad / 2)
    r_pad = int(total_pad / 2)
    pad = np.array([255] * 28).reshape((28,1))
    padded = np.concatenate([*repeat(pad, l_pad), a, *repeat(pad, r_pad)], axis=1).astype("uint8")
    return Image.fromarray(padded)
```
